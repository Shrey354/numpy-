```python
import numpy as np
```

```python
from time import process_time
```

```python
list = [i for i in range(10000)]
start_time = process_time()
list = [i*2 for i in list]
end_time = process_time()
print(end_time - start_time)
```

```python
np_array = np.array([i for i in range(10000)])
start_time = process_time()
np_array = np_array * 2
end_time = process_time()
print(end_time - start_time)
```

Array Manipulation


```python
array = np.random.randint(0,10,(3,4))
print(array)
print(array.shape)
```

```python
#Transpose
trans = np.transpose(array)
print(trans)
```

```python
print(trans.shape)
```

```python
trans2 = array.T
print(trans2)
```

```python
print(trans2.shape)
```

```python
#reshaping array
array = np.random.randint(0,10,(3,4))
print(array)
```

```python
reshape = array.reshape(4,3)
print(reshape)
```
